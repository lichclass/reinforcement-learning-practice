# Assignment 1: Bandits and Exploration

## Section 0: Preliminaries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Section 1: Problem Setup

- Define the multi-armed bandit problem with k arms (default: 10 arms)
- Assign each arm a true action value (randomly drawn from a normal distribution)
- Generate stochastic rewards for each action (from a normal distribution with mean = true action value)

```python
class Bandit():
    def __init__(self, k=10):
        self.k_arms = 10
        self.q_true = np.random.normal(0, 1, k)

    def get_reward(self, action):
        return np.random.normal(self.q_true[action], 1)

bandit = Bandit(k=10)
print("True action values:", bandit.q_true)
print("Reward from action 2:", bandit.get_reward(2))
```

## Section 2: Action Selection Strategies

- Implement ε-greedy action selection, where:
    - With probability ε, a random action is chosen (exploration).
    - With probability 1 - ε, the current best action is chosen (exploitation).
- Implement optimistic initial values, where we initialize Q-values optimistically to encourage exploration.
- Implement UCB (Upper Confidence Bound)

```python
class Agent():
    def __init__(self, k=10, epsilon=0.1, alpha=None, initial_q=0):
        self.k = k
        self.epsilon = epsilon
        self.alpha = alpha
        self.q_estimates = np.full(k, initial_q, dtype=float)
        self.action_counts = np.zeros(k, dtype=int)

    def select_action(self):
        r = np.random.rand()
        if r < self.epsilon:
            return np.random.randint(self.k)
        else:
            return np.argmax(self.q_estimates)

    def update_estimates(self, action, reward):
        self.action_counts[action] += 1
        
        if self.alpha is None:
            alpha = 1 / self.action_counts[action]
        else:
            alpha = self.alpha
        self.q_estimates[action] += alpha * (reward - self.q_estimates[action])

class UCB_Agent:
    def __init__(self, k=10, epsilon=2, alpha=None, initial_q=0):
        """Initialize agent with UCB action selection."""
        self.k = k
        self.epsilon = epsilon # Exploration parameter
        self.alpha = alpha
        self.q_estimates = np.full(k, initial_q, dtype=float)  # Q-values
        self.action_counts = np.zeros(k, dtype=int)  # Count of each action
        self.total_steps = 0  # Track total time steps

    def select_action(self):
        """Select an action using Upper Confidence Bound (UCB)."""
        self.total_steps += 1
        if 0 in self.action_counts:
            return np.argmin(self.action_counts)  # Pick untried action first
        
        # Compute UCB values
        ucb_values = self.q_estimates + self.epsilon * np.sqrt(np.log(self.total_steps) / self.action_counts)
        return np.argmax(ucb_values)

    def update_estimates(self, action, reward):
        """Update Q-value estimates."""
        self.action_counts[action] += 1
        
        if self.alpha is None:
            alpha = 1 / self.action_counts[action]  # Sample-average
        else:
            alpha = self.alpha  # Constant step size
        
        self.q_estimates[action] += alpha * (reward - self.q_estimates[action])
```

## Section 3: Running the Bandit Algorithm

- Simulate the bandit problem for N runs over T timesteps.
- Track average rewards and optimal action selection percentages over time.

```python
def run_bandit_experiment(k=10, time_steps=1000, epsilon=0.1, alpha=None, initial_q=0):
    bandit = Bandit(k)
    agent = Agent(k, epsilon, alpha, initial_q)
    rewards = np.zeros(time_steps)

    for t in range(time_steps):
        action = agent.select_action()
        reward = bandit.get_reward(action)
        agent.update_estimates(action, reward)
        rewards[t] = reward

    return rewards
```

```python
def run_ucb_experiment(k=10, time_steps=1000, epsilon=2, alpha=None, initial_q=0):
    """Run the UCB bandit algorithm and return reward history."""
    bandit = Bandit(k)
    agent = UCB_Agent(k, epsilon, alpha, initial_q)
    rewards = np.zeros(time_steps)

    for t in range(time_steps):
        action = agent.select_action()  # Select action using UCB
        reward = bandit.get_reward(action)  # Get reward
        agent.update_estimates(action, reward)  # Update Q-values
        rewards[t] = reward  # Store reward
    
    return rewards
```

## Section 4: Plotting and Analysis

- Plot average reward vs. time for different $\epsilon$ values
- Plot % of optimal action taken over time for different strategies

```python
# Run experiment
N = 1000  # Number of steps
np.random.seed(6)

greedy_rewards = run_bandit_experiment(k=10, time_steps=N, epsilon=0.1, alpha=0.1, initial_q=0)
optimistic_rewards = run_bandit_experiment(k=10, time_steps=N, epsilon=0.1, alpha=0.1, initial_q=5)
ucb_rewards = run_ucb_experiment(k=10, time_steps=N, epsilon=2, alpha=0.1, initial_q=0)

# Compute average rewards over time
greedy_avg_rewards = np.cumsum(greedy_rewards) / (np.arange(N) + 1)
optimistic_avg_rewards = np.cumsum(optimistic_rewards) / (np.arange(N) + 1)
ucb_avg_rewards = np.cumsum(ucb_rewards) / (np.arange(N) + 1)

# Plot both
plt.figure(figsize=(10, 5))
plt.plot(greedy_avg_rewards, label="ε-Greedy (initial_q=0)", linestyle="--")
plt.plot(optimistic_avg_rewards, label="Optimistic Initialization (initial_q=5)", linestyle="-")
plt.plot(ucb_avg_rewards, label="Upper Confidence Bound", linestyle=":")
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.title("Comparison: ε-Greedy vs. Optimistic Initialization vs. Upper Confidence Bound")
plt.legend()
plt.grid()
plt.show()
```
